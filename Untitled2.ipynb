When creating a custom exception in Python, it is recommended to inherit from the Exception class or one of its subclasses. Here's why using the Exception class as the base class for custom exceptions is beneficial:
1. Inherits Standard Exception Behavior
2. Consistent Exception Hierarchy
3. Identifies Exceptions as Errors
4. Facilitates Exception Handling

```python
def print_exception_hierarchy(exc_class, indent=0):
    print(' ' * indent + exc_class.__name__)
    for subclass in exc_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)
print_exception_hierarchy(BaseException)
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
Ans:
The base class for exceptions that happen during arithmetic operations in Python is called ArithmeticError. It serves as a common foundation for exceptions connected to arithmetic and is a subclass of the Exception class.

```python
try:
    result = 1 / 0
except ZeroDivisionError as e:
    print("Error: Division by zero occurred.")
```

```python
import sys
try:
    result = sys.maxsize + 1
except OverflowError as e:
    print("Error: Arithmetic operation resulted in overflow.")
```

Q5. Explain ImportError. What is ModuleNotFoundError?
Ans:
ImportError and ModuleNotFoundError are both exceptions that occur when importing modules in Python, but they have slight differences in their behavior and use cases.

```python
try:
    import cj_module
except ImportError:
    print("Failed to import the module.")
```

```python
try:
    import dj_module
except ModuleNotFoundError:
    print("The requested module could not be found.")
```
